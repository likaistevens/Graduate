```python
def J(theta,x_b,y):
    try:
        return np.sum((y - x_b.dot(theta))**2) / len(x_b)
    except:
        return float('inf')
```

```python
def dJ(theta,x_b,y):
    res = np.empty(len(theta))
    res[0] = np.sum(x_b.dot(theta) - y)
    for i in range(1, len(theta)):
        res[i] = (x_b.dot(theta) - y).dot(x_b[:,i])
    return res * 2 / len(x_b)
```

```python
def gradient_descent(x_b,y,initial_theta,eta,n_iters = 1e4,epsilon=1e-8):
    theta = initial_theta
    i_iter = 0
    while i_iter < n_iters:
        gradient = dJ(theta,x_b,y)
        last_theta = theta
        theta = theta - eta * gradient
        
        if(abs(J(theta,x_b,y) - J(last_theta,x_b,y)) < epsion):
            break
        i_iter += 1
    return theta
```

```python
import math
import numpy as np

def func_2d(x):
    return (50/x[0]) + x[1]*(x[0]-10)**2

def grad_2d(x):
    deriv0 = -50/(x[0]**2) + 2*x[1]*(x[0]-10)
    deriv1 = 2*x[1]*(x[0]-10)
    return np.array([deriv0, deriv1])

def gradient_descent_2d(grad, cur_x = np.array([0.1,0.1]),learning_rate = 0.01, precision = 0.0001, max_iters = 10000):
    print(f"{cur_x}as the initial parameter...function value:{func_2d(cur_x)}")
    for i in range(max_iters):
        grad_cur = grad(cur_x)
        if np.linalg.norm(grad_cur, ord=2) < precision:
            break
        cur_x = cur_x - grad_cur * learning_rate
        print(i," iteration：x:",cur_x,"function value:",func_2d(cur_x))
        
    print("minimizer x = ", cur_x)
    print("minimum of function",func_2d(cur_x))
    return cur_x

if __name__ =='__main__':
    gradient_descent_2d(grad_2d,cur_x=np.array([400,400]),learning_rate=0.001,precision=0.1,max_iters=10000)
```
